```python
countries = list(data[data['Country code'] < 900]['Destination'])
```

```python
usa_index = usa.index[0]
```

```python
usa_vals = [(data.loc[usa_index, c], c) for c in countries]
```

```python
usa_filtered = filter(lambda x: ~np.isnan(x[0]), usa_vals)
```

```python
usa_sorted = sorted(usa_filtered, reverse=True)
```
